```python
import numpy as np
import networkx as nx
import statistics
import math
import random
import os
from collections import Counter
import matplotlib.pyplot as plt
from time import time

import model, graphics
```

### Graph example <a class="anchor" id="chapter1"></a>

```python
# Set parameters
n = 50
m = 5
c0 = 0.8
c1 = 0.2
h = 0.8

# Run the experiment
G, node_class = model.model(n, m, c0, c1, h)

# Plot resulting graph
node_color = ['blue' if c == 0 else 'orange' for c in node_class]
node_size = np.array(list(dict(G.degree()).values())) * 5
nx.draw(G, node_color=node_color, node_size=node_size, width=0.5)
plt.show()
```

### Run small experiments <a class="anchor" id="chapter2"></a>

Small experiments for a better understanding of how to run experiments.

This section executes 20 simulations of 4 experiments. In the first experiment $h=0$, in the second one $h=0.3$, in the third one $h=0.7$ and in the last one $h=1$. In all cases, $n=500$ and $m=5$.

```python
# folder in which the data will be stored
current_path = os.path.abspath(os.getcwd()).replace('\\', '/')
current_path += '/outputs/test_data'
```

```python
# set the parameters of each experiment
n = 500
m = 5
c0 = [0.8, 0.8, 0.8, 0.8]
c1 = [0.2, 0.2, 0.2, 0.2]
h = [0, 0.3, 0.7, 1]
n_simulations = 20

# execute the experiments
for i in range(len(c0)):
    final_path = current_path + '/' + graphics.get_folder_name(n=n, m=m, c0=c0[i], c1=c1[i], h=h[i])
    model.compute_simulation(n=n, m=m, c0=c0[i], c1=c1[i], h=h[i],
                             n_simulations=n_simulations, folder_name=final_path)
```
